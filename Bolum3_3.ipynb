```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
from scipy import stats
import seaborn as sns
```

```python
dagilim = np.random.normal(2,4,1000)
```

```python
plt.figure(figsize=(4,2), dpi = 200)
count, bins, ignore = plt.hist(dagilim, 50, density=True)
plt.plot(bins, 1/(4*np.sqrt(2*np.pi))*np.exp(-(bins-2)**2/(2*4**2)),
        linewidth=3, color='y')
plt.show()
```

```python
dagilim.sort()

plt.figure(figsize=(4,2), dpi=200)
nsample = 100
res = stats.probplot(dagilim, plot=plt)
plt.title("Q-Q Grafiği")
plt.xlabel("Örnek Veriler")
plt.ylabel("Normal Dağılımlı Veriler")
plt.show()
```
